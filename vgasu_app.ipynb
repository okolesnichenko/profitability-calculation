```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model

months = ['Январь','Февраль','Март','Апрель','Май','Июнь','Июль','Август','Сентябрь','Октябрь','Ноябрь','Декабрь']
```

<h2> 2016 </h2>

```python
excel_file_2016 = "2016/Данные 2016.xlsx"

sales_profit = pd.read_excel(excel_file_2016, sheet_name = 'Прибыль от реализации', index_col=0).fillna(0)
operating_activities = pd.read_excel(excel_file_2016, sheet_name = 'Операционная деятельность', index_col=0).fillna(0)
nonsales_activities = pd.read_excel(excel_file_2016, sheet_name = 'Внереализационная деятельность', index_col=0).fillna(0)
tax = pd.read_excel(excel_file_2016, sheet_name = 'Налог на прибыль', index_col=0).fillna(0)
current_assets = pd.read_excel(excel_file_2016, sheet_name = 'Оборотные активы', index_col=0).fillna(0)
fixed_assets = pd.read_excel(excel_file_2016, sheet_name = 'Внеоборотные активы', index_col=0).fillna(0)
loans = pd.read_excel(excel_file_2016, sheet_name = 'Кредиты и займы', index_col=0).fillna(0)
accounts_payable = pd.read_excel(excel_file_2016, sheet_name = 'Кредиторская задолженность', index_col=0).fillna(0)
longterm_loans = pd.read_excel(excel_file_2016, sheet_name = 'Долгосрочные займы', index_col=0).fillna(0)
revenues_from_sales = pd.read_excel(excel_file_2016, sheet_name = 'Выручка от реализации', index_col=0).fillna(0)
equity = pd.read_excel(excel_file_2016, sheet_name = 'Собственный капитал', index_col=0).fillna(0)
```

<h3> Расчет чистой прибыли </h3>

```python
net_profit = sales_profit + operating_activities + nonsales_activities + tax
net_profit.index.name = "Чистая прибыль"
net_profit
```

<h3> Расчет рентабельности продаж </h3>

```python
return_on_sales = net_profit / revenues_from_sales
return_on_sales.index.name = "Рентабельность продаж"
return_on_sales
```

```python
# 0 = Ртищево, 1 = Лиски, 2 = Стойленская, 3 = Валуйки, 4 = Казинка, 5 = Кочетовка, 6 = Служба вагонного хозяйства
data = return_on_sales.iloc[6,:].values[0:12]
plt.figure(figsize=(9,5))
plt.bar(months, data)
plt.xlabel('Месяцы')
plt.ylabel('Служба вагонного хозяйства')
plt.show()
```

<h3> Расчет активов </h3>

```python
assets = current_assets + fixed_assets
assets.index.name = "Активы"
assets
```

<h3> Расчет оборачиваемости активов </h3>

```python
asset_turnover = revenues_from_sales / assets
asset_turnover.index.name = "Оборачиваемость активов"
asset_turnover
```

```python
# 0 = Ртищево, 1 = Лиски, 2 = Стойленская, 3 = Валуйки, 4 = Казинка, 5 = Кочетовка, 6 = Служба вагонного хозяйства
data = asset_turnover.iloc[6,:].values[0:12]
plt.figure(figsize=(9,5))
plt.bar(months, data) 
plt.xlabel('Месяцы')
plt.ylabel('Служба вагонного хозяйства')
plt.show()
```

<h3> Расчет рентабельности активов </h3>

```python
return_on_assets = return_on_sales * asset_turnover
return_on_assets.index.name = "Рентабельность активов"
return_on_assets
```

```python
# 0 = Ртищево, 1 = Лиски, 2 = Стойленская, 3 = Валуйки, 4 = Казинка, 5 = Кочетовка, 6 = Служба вагонного хозяйства
data = return_on_assets.iloc[6,:].values[0:12]
plt.figure(figsize=(9,5))
plt.bar(months, data) 
plt.xlabel('Месяцы')
plt.ylabel('Служба вагонного хозяйства')
plt.show()
```

<h3> Расчет привлеченного капитала </h3>

```python
raised_capital = loans + accounts_payable + longterm_loans
raised_capital.index.name = "Привлеченный капитал"
raised_capital
```

<h3> Расчет авансового капитала </h3>

```python
advance_capital = raised_capital + equity
advance_capital.index.name = "Авансовый капитал"
advance_capital
```

<h3> Расчет финансового рычага </h3>

```python
financial_leverage = advance_capital / equity
financial_leverage.index.name = "Финансовый рычаг"
financial_leverage
```

```python
# 0 = Ртищево, 1 = Лиски, 2 = Стойленская, 3 = Валуйки, 4 = Казинка, 5 = Кочетовка, 6 = Служба вагонного хозяйства
data = financial_leverage.iloc[6,:].values[0:12]
plt.figure(figsize=(9,5))
plt.bar(months, data) 
plt.xlabel('Месяцы')
plt.ylabel('Служба вагонного хозяйства')
plt.show()
```

```python
return_on_equity = return_on_assets * financial_leverage
return_on_equity.index.name = "Рентабельность собственного капитала"
#return_on_equity['2016 год'].values
return_on_equity
```

<h2> Прогноз показателя "Служба вагонного хозяйства"  </h2>

```python
length = 36

y = np.concatenate([data2016, data2017, data2018])
x = np.array(range(1,37))

x = x.reshape(length, 1)
y = y.reshape(length, 1)

regr = linear_model.LinearRegression()
regr.fit(x, y)

plt.scatter(x, y,  color='black')
plt.plot(x, regr.predict(x), color='blue', linewidth=3)
plt.xticks(())
plt.yticks(())
plt.show()

Z = np.array(range(37, 49))
Z = Z.reshape(len(Z), 1)

count = 0
prediction2019 = regr.predict(Z)
print(2019)
for tmp in prediction2019:
    str = '{month} {tmp}'.format(month = months[count], tmp = tmp[0])
    print(str)
    count+=1
```
